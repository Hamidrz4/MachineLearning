```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_moons
from sklearn.datasets import make_circles
```

```python
# Data Preparation (Iris)
iris = datasets.load_iris()
x = iris.data[:,2:4]
y = iris.target
```

```python
# Try different kernels
model1 = SVC(kernel="linear", C=10, gamma="auto").fit(x,y)
model2 = SVC(kernel="rbf", C=10, gamma="auto").fit(x,y)
model3 = SVC(kernel="poly", C=10, gamma="auto").fit(x,y)
```

```python
# Using subplots & Plot_decision_regions
figure, ax = plt.subplots(1,3)
plot_decision_regions(x, y, clf=model1, legend=2,ax=ax[0]);
plot_decision_regions(x, y, clf=model2, legend=2,ax=ax[1]);
plot_decision_regions(x, y, clf=model3, legend=2,ax=ax[2]);
```
